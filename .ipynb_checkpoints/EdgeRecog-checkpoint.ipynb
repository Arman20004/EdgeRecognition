```python
#importing libraries

import cv2
import os
from matplotlib import pyplot as plt
```

```python
# File path to image 

img_path = os.path.join(os.getcwd(), "EdgeRecognition", "Sample", "Img (3).JPG")
```

```python
# Reading image

image = cv2.imread(img_path)
```

```python
# If image is opened correctly you should see it, otherwise you will see its type
if (image is None):
    print("Expected image got")
    print(type(image))
else:
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.show()
```

```python
# Turn image to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
# Apply some blur
blur = cv2.GaussianBlur(gray, (5, 5), 0)
```

```python
# Getting edges with canny
canny = cv2.Canny(blur, threshold1 = 40, threshold2 = 45)
```

```python
# Blured grayscale image
plt.imshow(cv2.cvtColor(blur, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```python
# Edges of an object

plt.imshow(cv2.cvtColor(canny, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```
